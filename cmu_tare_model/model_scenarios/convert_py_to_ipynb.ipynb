# Convert .py File to Jupyter Notebook
Use the `jupytext` command-line tool to convert a Python script (`.py` file) into a Jupyter notebook (`.ipynb` file).

```python
# Convert a Python script to a Jupyter notebook using `jupytext`
!jupytext --to notebook "C:\Users\14128\Research\cmu-tare-model\cmu_tare_model\model_scenarios\tare_baseline_v2.1.py"
```

# Verify the Conversion
Check the newly created `.ipynb` file to ensure it contains the expected content from the `.py` file.

```python
# Verify the Conversion

# Import the `nbformat` library to read the `.ipynb` file
import nbformat

# Load the newly created Jupyter notebook file
notebook_path = "tare_baseline_v2.1.ipynb"
with open(notebook_path, "r", encoding="utf-8") as f:
    notebook_content = nbformat.read(f, as_version=4)

# Display the content of the first cell to verify the conversion
first_cell = notebook_content['cells'][0]
print("First cell content:")
print(first_cell['source'])

# Check the total number of cells in the notebook
total_cells = len(notebook_content['cells'])
print(f"\nTotal number of cells in the notebook: {total_cells}")
```
