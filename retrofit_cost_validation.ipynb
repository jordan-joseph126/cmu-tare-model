```python
import seaborn as sns
import matplotlib.pyplot as plt

# Abbreviate the names in the 'upgrade_package' column
abbreviations = {
    "conventional envelope": "Conventional",
    "IECC envelope": "IECC",
    "Phius envelope": "Phius"
}
df_abc_retrofit_costs['upgrade_package'] = df_abc_retrofit_costs['upgrade_package'].apply(lambda x: x.split(' - ')[0] + ' - ' + abbreviations[x.split(' - ')[1].split(' + ')[1]])

sns.set_theme(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15, 6))

# Plot the high median incremental costs
sns.set_color_codes("pastel")
sns.barplot(x="upgrade_package", y="incremental_costs_high", data=df_abc_retrofit_costs,
            label="High - Median Incremental Project Costs", color="b")

# Plot the low median incremental costs
sns.set_color_codes("muted")
sns.barplot(x="upgrade_package", y="incremental_costs_low", data=df_abc_retrofit_costs,
            label="Low - Median Incremental Project Costs", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 24), ylabel="Incremental (Net Equipment) Project Costs [$USD-2021]",
       xlabel="")
sns.despine(left=True, bottom=True)

# Rotate x-axis labels
plt.xticks(rotation=90)

plt.show()
```

```python
def process_upgrade_package(entry):
    try:
        # Splitting by ' - ' and ensuring there are enough parts
        parts = entry.split(' - ')
        if len(parts) < 2 or '+' not in parts[1]:
            # If the format is not as expected, return the original or a default formatted entry
            return "Format Error - Check Entry"
        
        # Using the abbreviation dictionary safely
        sub_parts = parts[1].split(' + ')
        sub_part_final = sub_parts[1] if len(sub_parts) > 1 else sub_parts[0]
        return parts[0] + ' - ' + abbreviations.get(sub_part_final, "Unknown")
    except Exception as e:
        # Handle any unexpected errors
        return "Error - Check Entry"

# Apply the function to the dataframe column
df_abc_retrofit_costs['upgrade_package'] = df_abc_retrofit_costs['upgrade_package'].apply(process_upgrade_package)
df_abc_retrofit_costs
```
